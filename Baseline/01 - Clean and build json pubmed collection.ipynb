```python

from utils import to_date, dir_tree_run, multiprocess_xml_read, multiprocess_xml_to_json, collection_iterator, f_recall, create_filter_query_function
import json
from lxml import etree
import os
import gc
import sys
import pickle
import tempfile
import shutil
import tarfile

from functools import reduce


baseline_name = "baseline_2020"
index_name = "bioasq_9b"
```

```python

# xml baseline files names

articles_tar_names = []
dir_tree_run(lambda x:articles_tar_names.append(x), "/backup/pubmed_archive/"+baseline_name)
articles_tar_names = sorted(articles_tar_names)

# convert xml articles to json format
zipped_collection = "/backup/pubmed_archive_json/"+index_name+".tar.gz"
# check if tar.gz exists first
if not os.path.exists(zipped_collection):
    articles = multiprocess_xml_read(articles_tar_names, n_process=20)
```

```python
print("Size articles before duplicated remove", len(articles))

unique_articles = {}



no_data_skips = 0
small_data_skips = 0
    
for x in articles:
    
    # skip articles that do not contain data information
    if x["pubdate"]=="" or x["pubdate"]==None:
        no_data_skips+=1
        continue
    
    # this will override base on the data time
    # TODO test with > signals instead
    if x["id"] in unique_articles and to_date(x["pubdate"])<to_date(unique_articles[x["id"]]["pubdate"]):
        small_data_skips+=1
        continue

    unique_articles[x["id"]] = x
    
print("No data skips", no_data_skips)
print("Small data skips", small_data_skips)
print("Size articles after duplicated remove", len(unique_articles))
```

```python

# batch save
unique_articles = list(unique_articles.values())
size = len(unique_articles)
print(size)
itter = int(3e6)

f_names=[]
temp_dir = tempfile.mkdtemp()
        
try:
    for i in range(0, size, itter):
        file_name = "{0}/{1}_{2:08}_to_{3:08}".format(temp_dir,index_name,i, min(size, i+itter))
        print("Save file",file_name,":",end="")
        json.dump(unique_articles[i:i+itter], open(file_name,"w"))
        f_names.append(file_name)
        print("Done")
    
    print("Start the compression")
    # build tar file
    with tarfile.open(zipped_collection, "w:gz") as tar:
        for name in f_names:
            tar.add(name)

except Exception as e:
    raise e # maybe handle the exception in the future
finally:
    # always remove the temp directory
    print("Remove {}".format(temp_dir))
    shutil.rmtree(temp_dir)
```

```python
def p_open_xml(proc_id, xml_files, output_dir):
    import pubmed_parser as pp
    
    def filter_mesh(string):
        return " ".join(map(lambda y:y[0], map(lambda x: x.split(";"), string.split(":")[1:])))
    
    print("[Process-{}] Started".format(proc_id))
    articles = []
    for file_name in xml_files:
        print(proc_id, file_name)
        try:
            articles.extend(pp.parse_medline_xml(file_name, year_info_only=False, nlm_category=False))
        except etree.XMLSyntaxError:
            print("Error on File " + file_name)
        
        gc.collect()
            
    articles_mapped = list(map(lambda x:{"id":x["pmid"],
                                         "title":x["title"],
                                         "abstract":x["abstract"],
                                         "keywords":x["keywords"],
                                         "pubdate":x["pubdate"],
                                         "mesh_terms":filter_mesh(x["mesh_terms"]),
                                         "delete":x["delete"]}
                               ,articles))

    file_name = output_dir+"/pubmed_2019_{0:03}.p".format(proc_id)
    print("[Process-{}]: Store {}".format(proc_id, file_name))

    with open(file_name, "wb") as f:
        pickle.dump(articles_mapped, f)

    del articles
    print("[Process-{}] Ended".format(proc_id))


articles = multiprocess_xml_read(articles_tar_names, n_process=12,open_fn=p_open_xml)
```
