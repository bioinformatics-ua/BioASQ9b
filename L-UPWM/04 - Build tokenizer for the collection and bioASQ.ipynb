```python
from nir.tokenizers import Regex, BioCleanTokenizer, BioCleanTokenizer2, Regex2
from nir.utils import plot_tokenizer_word_frequency
from utils import collection_iterator
import json
import os

cache_folder = "/backup/NIR_BioASQ"

index_name = "bioasq_9b"

zipped_collection = "/backup/pubmed_archive_json/"+index_name+".tar.gz"
```

```python
#_class = Regex
#_class = Regex2
_class = BioCleanTokenizer
#_class = BioCleanTokenizer2

tk = _class(cache_folder=os.path.join(cache_folder, "tokenizers"),
            prefix_name=index_name,
            min_word_frequency=5,
            n_process=12)
```

```python
    
# fit on corpus
tk.fit_tokenizer_multiprocess(collection_iterator(zipped_collection, f_map=lambda x:x["title"]+" "+x["abstract"]))
```

```python
with open("training9b.json", "r") as f:
    data_to_tokenize = list(map(lambda x:x["body"], json.load(f)["questions"]))

tk.fit_on_texts(data_to_tokenize)
```

```python
plot_tokenizer_word_frequency(tk)
```
